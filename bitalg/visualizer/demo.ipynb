```python
from bitalg.visualizer.main import Visualizer
```

```python
import math
from math import pi

r = 50
n = 10
points = [(math.cos(2*pi/n*x)*r,math.sin(2*pi/n*x)*r) for x in range(0,n+1)]


vis = Visualizer()

for p in points:
    vis.add_point(p[0], p[1], color='blue', linewidth=5)

for p in points:
    vis.add_point(p[0] + 10, p[1], color='orange', linewidth=0.5)


vis.add_polygon(points, color='green', alpha=0.4)


for p in points:
    vis.add_line_segment((p[0], p[1]), (p[0] + 10, p[1]), color='red', linewidth=2)


vis.show()
```

```python
vis.show_gif()
```
